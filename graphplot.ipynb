```python
import matplotlib.pyplot as plt
            
```

```python
x=[100,200]
y=[23,64]
y1=[150,176]
x1=[120,144]
plt.plot(x1,y1,label="road")
plt.plot(x,y,label="railway")
plt.xlabel("score")
plt.ylabel("names")
plt.legend()
plt.show()
```

```python
x=["mudit","sanyam","shoubhik"]
y=[4,3,2]
plt.bar(x,y,label="Theory")
plt.grid(color='yellow')
plt.xlabel("names")
plt.ylabel("marks")
plt.yscale('linear')
plt.legend()
plt.show()
```

```python
x=[22,30,55,66]
y=[2,44,55,6]
x1=[13,23]
y1=[34,21]
plt.scatter(x,y,label="well",marker='x',s=50)
plt.scatter(x1,y1,label="water",s=50)
plt.legend()
plt.grid(color='g')
plt.show()
```
